#### Write a function to get the number of jobs for the Python technology.<br>
> Note: While using the lab you need to pass the **payload** information for the **params** attribute in the form of **key** **value** pairs.
  Refer the ungraded **rest api lab** in the course **Python for Data Science, AI & Development**  <a href="https://www.coursera.org/learn/python-for-applied-data-science-ai/ungradedLti/P6sW8/hands-on-lab-access-rest-apis-request-http?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDA0321ENSkillsNetwork928-2022-01-01">link</a>
  
 ##### The keys in the json are 
 * Job Title
 
 * Job Experience Required
 
 * Key Skills
 
 * Role Category
 
 * Location
 
 * Functional Area
 
 * Industry
 
 * Role 
 
You can also view  the json file contents  from the following <a href = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json">json</a> URL.


```python
def get_number_of_jobs_T(technology):
    # Define the API URL
    api_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"
    
    # Define the payload with parameters for locations and technology
    payload = {
        'locations': ['Los Angeles', 'New York', 'San Francisco', 'Washington DC', 'Seattle', 'Austin', 'Detroit'],
        'technology': technology,
    }

    # Make an API request with the payload
    response = requests.get(api_url, params=payload)

    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()
        
        # Extract the number of job postings for the specified technology
        number_of_jobs = 0
        for job in data:
            if job.get('Key Skills', '').lower().find(technology.lower()) != -1:
                number_of_jobs += 1
        
        return technology, number_of_jobs
    else:
        return technology, 0  # Return 0 if the request was not successful
```

```python
import requests
import openpyxl

# Define the API URL
api_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"

# Define the list of locations and technologies
locations = ['Los Angeles', 'New York', 'San Francisco', 'Washington DC', 'Seattle', 'Austin', 'Detroit']
technologies = ['C', 'C#', 'C++', 'Java', 'JavaScript', 'Python', 'Scala', 'Oracle', 'SQL Server', 'MySQL Server', 'PostgreSQL', 'MongoDB']

# Create a new Excel workbook
wb = openpyxl.Workbook()
ws = wb.active

# Create a dictionary to store the technology counts
technology_counts = {}

def get_number_of_jobs_L(location, technology):
    # Define the payload with the location parameter
    payload = {
        'Location': location,
    }

    # Make an API request with the payload
    response = requests.get(api_url, params=payload)

    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()

        # Extract the number of job postings for the specified technology
        number_of_jobs = 0
        for job in data:
            if job.get('Key Skills', '').lower().find(technology.lower()) != -1:
                number_of_jobs += 1
        
        # Store the count in the dictionary
        technology_counts[technology] = technology_counts.get(technology, 0) + number_of_jobs

        return location, technology, number_of_jobs
    else:
        return location, technology, 0  # Return 0 if the request was not successful
    
ws.append(['Location','Technology', 'Count'])

# Loop through locations and technologies, call the API, and write results to the Excel sheet
# Loop through locations and technologies
for location in locations:
    for technology in technologies:
        job_count = get_number_of_jobs_L(location, technology)  # Pass both location and technology
        ws.append([location, technology, job_count[2]])  # Append job_count[2] (the job count) to the worksheet

# Save the results to an Excel file on your local PC
local_file_path = "job-postings2.xlsx"  # Specify the local file path
wb.save(local_file_path)

# Print the technology counts
for tech, count in technology_counts.items():
    print(f"Technology: {tech}, Count: {count}")

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the job postings data from the CSV file
df = pd.read_excel("job-postings2.xlsx")

# Sort the data by salary in descending order
sorted_df = df.sort_values(by="Count", ascending=False)

# Create a bar chart
plt.figure()
plt.barh(sorted_df["Technology"], sorted_df["Count"], color='skyblue')
plt.xlabel("Count")
plt.ylabel("Technology")
plt.title("Popular Technology by count")
plt.gca().invert_yaxis()  # Invert the y-axis to display the highest salary at the top
plt.show()
```
