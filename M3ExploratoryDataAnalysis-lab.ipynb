## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
plt.figure(figsize=(10, 6))
df.boxplot(column='ConvertedComp')
plt.title('Box Plot for ConvertedComp')
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outliers = df[(df['ConvertedComp'] < lower_bound) | (df['ConvertedComp'] > upper_bound)]
num_outliers = len(outliers)
num_outliers
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.


```python
# your code goes here
# Calculate the correlation using the corr() method
correlation_age_comp_total = df['Age'].corr(df['CompTotal'])
correlation_age_code_rev_hrs = df['Age'].corr(df['CodeRevHrs'])
correlation_age_respondent = df['Age'].corr(df['Respondent'])
correlation_age_convertedcomp = df['Age'].corr(df['ConvertedComp'])

print("Correlation of Age with CompTotal:", correlation_age_comp_total)
print("Correlation of Age with CodeRevHrs:", correlation_age_code_rev_hrs)
print("Correlation of Age with Respondent:", correlation_age_respondent)
print("Correlation of Age with WorkWeekHrs:", correlation_age_work_week_hrs)
print("Correlation of Age with ConvertedComp:", correlation_age_convertedcomp)


```
