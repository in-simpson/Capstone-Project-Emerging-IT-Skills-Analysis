# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt

# Query to retrieve the data
QUERY = """
SELECT ConvertedComp
FROM master
"""

# Read the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Plot the histogram
plt.hist(df['ConvertedComp'], bins=30, edgecolor='k')
plt.xlabel('Salary (ConvertedComp)')
plt.ylabel('Frequency')
plt.title('Histogram of ConvertedComp')
plt.show()

```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
# Plot a box plot of Age
QUERY = """
SELECT Age
FROM master
"""

# Read the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

plt.boxplot(df['Age'].dropna())
plt.ylabel('Age')
plt.title('Box Plot of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT Age, workWeekHrs
FROM master
"""

# Read the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)
# Create a scatter plot of Age and WorkWeekHrs
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs,CodeRevHrs,Age
FROM master
"""

# Read the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Create a bubble plot of WorkWeekHrs and CodeRevHrs with Age as bubble size
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age'], alpha=0.5)
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs (Age as bubble size)')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

# Query to retrieve the top 5 databases respondents wish to learn next year
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""
df = pd.read_sql_query(QUERY, conn)

# Create a pie chart
plt.pie(df['Count'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%')
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here


# Query to retrieve median WorkWeekHrs and CodeRevHrs for the age group 30 to 35
QUERY = """
SELECT Age, AVG(WorkWeekHrs) as Median_WorkWeekHrs, AVG(CodeRevHrs) as Median_CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
GROUP BY Age
ORDER BY Age
"""
df = pd.read_sql_query(QUERY, conn)

# Create a stacked chart
plt.bar(df['Age'], df['Median_WorkWeekHrs'], label='WorkWeekHrs', alpha=0.7)
plt.bar(df['Age'], df['Median_CodeRevHrs'], label='CodeRevHrs', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here


# Query to retrieve the median ConvertedComp for all ages from 45 to 60
QUERY = """
SELECT Age, AVG(ConvertedComp) as Median_ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
GROUP BY Age
ORDER BY Age
"""
df = pd.read_sql_query(QUERY, conn)

# Create a line chart
plt.plot(df['Age'], df['Median_ConvertedComp'])
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Line Chart of Median ConvertedComp for Ages 45 to 60')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here


# Query to retrieve the counts for each MainBranch category
QUERY = """
SELECT MainBranch, COUNT(*) as Count
FROM master
GROUP BY MainBranch
"""
df = pd.read_sql_query(QUERY, conn)

# Create a horizontal bar chart
plt.barh(df['MainBranch'], df['Count'])
plt.xlabel('Count')
plt.title('Horizontal Bar Chart of MainBranch')
plt.show()
```
